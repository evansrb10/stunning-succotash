```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
# Read the City and Ride Data
city_df = pd.read_csv("data/city_data.csv")
ride_df = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset
combined_pyber_data = pd.merge(ride_df, city_df, on="city")
#combined_pyber_data = combined_pyber_data.set_index("city")
# Display the data table for preview
combined_pyber_data.head()
```

```python
#grouping merged data by city
ride_city = combined_pyber_data.groupby("city")

#average fare by city
avg_fare = ride_city.mean()["fare"]

#total number of rides by city
total_rides = ride_city.count()["ride_id"]

#total number of drivers by city
total_drivers = ride_city.mean()["driver_count"]

#define types of cities
city_type = city_df.set_index("city")["type"]

# Obtain the x and y coordinates for each of the three city types
#create dataframe for all values
city_info = pd.DataFrame({
    "Average Fare by City": avg_fare,
    "Rides by City": total_rides,
    "Drivers by City": total_drivers,
    "Types of City": city_type
})

city_info.head()
```

## Bubble Plot of Ride Sharing Data

```python
# dataframe by city type
urban = city_info[city_info['Types of City'] == 'Urban']
suburban = city_info[city_info['Types of City'] == 'Suburban']
rural = city_info[city_info['Types of City'] == 'Rural']

#Build the scatter plots for each city types
# Incorporate a text label regarding circle size
plt.scatter(urban["Rides by City"], urban["Average Fare by City"], s=urban["Drivers by City"]*15, alpha =.7,
                  color = 'lightcoral', edgecolor = 'maroon', linewidth=2, label = 'Urban')
plt.scatter(suburban["Rides by City"], suburban["Average Fare by City"], s=suburban["Drivers by City"]*15, alpha =.5,
                  color = 'lightskyblue', edgecolor = 'navy', linewidth=2, label = 'Suburban')
plt.scatter(rural["Rides by City"], rural["Average Fare by City"], s=rural["Drivers by City"]*15, alpha =.3,
                  color = 'gold', edgecolor = 'black', linewidth=2, label = 'Rural')
#F08080
# Incorporate the other graph properties
#Add Axis and Title
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Pyber Ride Sharing Data (2019)')
plt.figtext(1,.5,'Note: Circle size correlates with driver count per city')

# Create a legend
#Other Plot Details
plt.grid()
lgnd = plt.legend()
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]

# Save Figure
plt.savefig("../Images/bubble.png")

# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
#group by city type
type_group = combined_pyber_data.groupby('type')

#all fares and totals
total_fares = type_group['fare'].sum()
rural_total = total_fares['Rural']
urban_total = total_fares['Urban']
suburban_total = total_fares['Suburban']


# Build Pie Chart
labels = ['Rural', 'Urban', 'Suburban']
sums = [rural_total, urban_total, suburban_total]
colors = ['gold', 'lightcoral','lightskyblue']
explode = (0, .2, 0)
plt.pie(sums, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 270, explode = explode)
plt.title('% of Total Fares by City Type')

# Save Figure
plt.savefig("../Images/fares.png")
```

## Total Rides by City Type

```python
# Calculate Ride Percents

#all fares and totals
total_rides_sum = type_group['ride_id'].count()
rural_total = total_rides_sum['Rural']
urban_total = total_rides_sum['Urban']
suburban_total = total_rides_sum['Suburban']


# Build Pie Chart
labels = ['Rural', 'Urban', 'Suburban']
sums = [rural_total, urban_total, suburban_total]
colors = ['gold', 'lightcoral','lightskyblue']
explode = (0, .2, 0)
plt.pie(sums, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 240, explode = explode)
plt.title('% of Total RIdes by City Type')

# Save Figure
plt.savefig("../Images/rides.png")
```

```python
# Calculate Driver Percents
type_group_city = city_df.groupby('type')

#all fares and totals
total_drivers_sum = type_group_city['driver_count'].sum()
rural_total = total_drivers_sum['Rural']
urban_total = total_drivers_sum['Urban']
suburban_total = total_drivers_sum['Suburban']


# Build Pie Chart
labels = ['Rural', 'Urban', 'Suburban']
sums = [rural_total, urban_total, suburban_total]
colors = ['gold', 'lightcoral','lightskyblue']
explode = (0, .2, 0)
plt.pie(sums, labels = labels, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 240, explode = explode)
plt.title('% of Total Drivers by City Type')

# Save Figure
plt.savefig("../Images/drivers.png")
```
